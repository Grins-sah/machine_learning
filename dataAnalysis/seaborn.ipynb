```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
### line plot 
sns.lineplot(x="size",y="total_bill",data=tips)
plt.show()
```

```python
sns.pairplot(data=tips)
```
