```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [1,2,4,5,6,9]
## create a line plot
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line in matplotlib")
```

```python
x = [1,2,3,5,6,8]
y = [1,2,58,96,1,3]
plt.plot(x,y,color="green",linestyle="--",marker="o")
plt.grid(True)
```

```python
### pi chart
labels =['A','B','C',"D"]
sizes = [45,20,50,97]
explods = [0.2,0,0,0]
plt.pie(sizes,explode=explods,labels=labels,autopct="%1.1f%%")

```
