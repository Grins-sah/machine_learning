```python
import pandas as pd
import numpy as ny
import seaborn as sns
import matplotlib.pylab as plt
df = pd.read_excel("02. flight-price.xlsx")
df
```

```python
df.info()
```

```python
df.head()
```

```python
### date => day month year
```

```python
df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df.info()
df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_time_hour'] = df['Arrival_Time'].str.split(" ").str[0].str.split(":").str[0]
df.drop(columns='Arrival_time_min',axis=1,inplace=True)
df
```

```python
df["Arrival_time_min"]  = df['Arrival_Time'].str.split(" ").str[0].str.split(":").str[1]
df['Arrival_time_hour']=df['Arrival_time_hour'].astype(int)
df["Arrival_time_min"]=df['Arrival_time_min'].astype(int)
print(df.head())
```

```python
df["departure_hour"] = df["Dep_Time"].str.split(":").str[0];
df["departure_min"] = df["Dep_Time"].str.split(":").str[1];
df["departure_hour"] = df['departure_hour'].astype(int);
df["departure_min"] = df['departure_min'].astype(int);
df.info()
```

```python
print(df.head())
```

```python
df.drop("Dep_Time",axis=1,inplace=True)
print(df.head())
```

```python
df['Airline'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Airline']]).toarray()
encoded
encoded_name = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
updated_df = pd.concat([df,encoded_name],axis=1)
```
