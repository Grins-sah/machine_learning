#### `logical reg`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
df = pd.DataFrame(x,columns=np.arange(10))
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logic_reg = LogisticRegression()
```

```python
logic_reg.fit(x_train,y_train)
```

```python
y_pred = logic_reg.predict(x_test)
```

```python
logic_reg.predict_proba(x_test)[:,0]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

`Hyper Parameter`

```python
model = LogisticRegression()
penalty = ['ll','l2','elasticnet']
c_val = [100,10,1.0,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag','saga']
```

```python
params = dict(penalty = penalty,C= c_val,solver = solver)
```

```python
### grid search cv
```

```python
from sklearn.model_selection import StratifiedGroupKFold
cv = StratifiedGroupKFold(n_splits=5)
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5)

from sklearn.model_selection import GridSearchCV
params = dict(penalty=['l1', 'l2', 'elasticnet', 'none'], C=[100, 10, 1.0, 0.01], solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'])
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
grid.fit(x_train, y_train)
```

```python
x_train
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

`Randomize CV`

```python
from sklearn.model_selection import RandomizedSearchCV
Random = RandomizedSearchCV(model,param_distributions=params,scoring='accuracy',cv=5,n_jobs=-1)
```

```python
Random.fit(x_train,y_train)
```

```python
Random.best_score_
```

```python
Random.best_params_
```

```python
y_pred = Random.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
