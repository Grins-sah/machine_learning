```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
df = pd.DataFrame({
    "color":["red","blue","green","red","blue","green","red","blue","green"]
})
df.head(10)
```

```python
encoder = OneHotEncoder();
encoded  = encoder.fit_transform(df[['color']]).toarray()
df_endoded = pd.DataFrame(encoded,columns=encoder.get_feature_names_out());
df_endoded
```

```python
pd.concat([df,df_endoded],axis=1)
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
encoder = OneHotEncoder()
encoded = encoder.fit_transform(tips[['sex']]).toarray()
df_encoded_sex = pd.DataFrame(encoded,columns=encoder.get_feature_names_out());
df_encoded_sex
pd.concat([tips,df_encoded_sex],axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({
    "size":["small","medium","large","small","medium","large","small","medium","large"]
})
df
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['small','medium','large']])
encoded = ordinal_encoder.fit_transform(df[['size']])
encoded
df1 = pd.DataFrame(encoded,columns=ordinal_encoder.get_feature_names_out())
pd.concat([df,df1],axis=1)
```
