```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
x,y = make_classification(
    n_samples= 1000,
    n_features=7,
    n_classes=3,
    n_informative=3,
    n_redundant=3,
    random_state=42

)
```

```python
x[:,0]
```

```python
sns.scatterplot(x=x[:,0],y = x[:,1],hue=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
classsifier = KNeighborsClassifier()
classsifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_true=y_test,y_pred=classsifier.predict(x_test)))
```

```python
ans = {"val":0}
for i in range(1,50):
    classsifier = KNeighborsClassifier(n_neighbors=i,p=2)
    classsifier.fit(x_train,y_train)
    ans["val"] = max(ans["val"],accuracy_score(y_true=y_test,y_pred=classsifier.predict(x_test)))
print(ans["val"])
    
```

```python
sns.scatterplot(x=x_test[:,0],y=x_test[:,1],hue=classsifier.predict(x_test))
```
