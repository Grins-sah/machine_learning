```python
def square(n):
    for i in range(n):
        yield i**2
```

```python
square(5)
```

```python
for i in square(3):
    print(i)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
print(next(gen))
```

### reading large files 
use to read large file

```python
def read_large(file_name):
    with open(file_name,'r') as file:
        for line in file:
            yield line
read = read_large("grins.txt");
read
```

```python
for i in read:
    print(i.strip());
```
