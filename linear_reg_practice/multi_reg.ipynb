```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
df = fetch_california_housing()
```

```python
df = pd.DataFrame(df.data,columns=df.feature_names)
```

```python
df
```

```python
df['price'] = fetch_california_housing().target
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
x = df.iloc[:,0:8]
y = df.iloc[:,8:]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
y_pred = reg.predict(x_test)
```

```python
# r score
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

```python
plt.scatter(y_test, y_pred, c='blue')  # Change 'blue' to any color you like
plt.xlabel("Actual Values (y_test)")
plt.ylabel("Predicted Values (y_pred)")
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
residule = y_test-y_pred
```

```python
sns.displot(residule,kind='kde')
```

```python
plt.scatter(y_pred,residule)
```

```python
import pickle
pickle.dump(reg,open('regressor.pkl','wb'))
```

```python
model= pickle.load(open('regressor.pkl','rb'))
```
